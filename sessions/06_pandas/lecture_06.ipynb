## DataFrame（データフレーム）を作ってみよう

Pandasを使えば，文字と数字が混ざっていてもエラーにならず，
それぞれのデータを正しく保ったまま分析ができるようになる．
まずは，先ほどの「リストのリスト」を，Pandasの魔法の箱 **`DataFrame（データフレーム）`** に入れてみよう．

`DataFrame` とは，皆さんがよく知っている **「Excelのシート（2次元の表）」** そのものである．表を作るついでに，列の名前（見出し）もつけてあげる．

````
import pandas as pd
# DataFrameに変換しつつ，列の名前（columns）を指定する
df = pd.DataFrame(data, columns=['名前', '睡眠タイプ', 'お手伝い時間(分)'])

# 表を表示する
display(df) # printでも表示されるがdisplayのほうが綺麗に表示される
````

実行すると，左端に行番号（インデックス）がつき，上に列名（カラム）がついた，
綺麗な「表」が出力される．これがデータ分析の基本となる形である．

最初の5行だけ取り出すときは`head()`を
最後の5行だけ取り出すときは`tail()`を用いる.

````
display(df.head())
display(df.tail())
````

## 表から1列だけを取り出す（Seriesの概念）

実務のデータ分析では，「全体の表から，特定の列だけを取り出して調べる」という操作を頻繁に行う．
先ほど作った表（`df`）から，「睡眠タイプ」の列だけを縦にスパッと切り出してみよう．取り出すときは `df['列の名前']` と書く．

````
# '睡眠タイプ'の列だけを取り出す
type_data = df['睡眠タイプ']
display(type_data)
````

このように，DataFrameから1列だけを切り出した縦長のデータのことを，Pandasでは **`Series（シリーズ）`** と呼ぶ．
つまり，**DataFrameとはいくつかのSeriesを横に並べて束ねたもの**である．

## 条件に合うデータだけを抽出する

実務では特定の条件を満たすデータだけを取り出すという操作をよく行う．
チーム編成を考えるために，「ぐっすり」タイプのポケモンだけを抽出してみよう．

````
# 「ぐっすり」タイプのポケモンだけを抽出する
gussuri_pokemon = df[df['睡眠タイプ'] == 'ぐっすり']
display(gussuri_pokemon)
````

**【解説：なぜこれで抽出できるのか？】**
内側から順番に見ていく．
1. `df['睡眠タイプ'] == 'ぐっすり'` を実行すると，「条件を満たしているか（TrueかFalseか）」のリスト（Series）が作られる．
2. その True / False のリストを外側の `df[ ]` に渡すことで，True となった行だけが残る仕組みになっている．

### よくあるエラー: 複数条件を書くときの鉄則！

複数条件を組み合わせるとき、Pythonの基礎で学んだ `and` や `or` という文字を使おうとすると、**ValueError**が出てプログラムが止まってしまう。

Pandasで複数条件を書くときは、絶対に守らなければならない「2つの掟」がある。

**掟1： `and` は `&` に、 `or` は `|` に置き換えるべし！**

**掟2： それぞれの条件式を、必ず `( )` で囲むべし！**

**やってはいけない書き方（エラーになる）**

````
# and を使っている ＆ カッコがない
hayai_suyasuya = df[df['お手伝い時間(分)'] < 50 and df['睡眠タイプ'] == 'すやすや']
````

**正しい書き方**

````
# & を使い、各条件を ( ) で囲んでいる
hayai_suyasuya = df[(df['お手伝い時間(分)'] < 50) & (df['睡眠タイプ'] == 'すやすや')]
````

※なぜカッコが必要なのか？
実は `&` や `|` は、掛け算や割り算と同じように「計算の優先順位」が非常に高い。カッコをつけないと、プログラムが `50 & df['睡眠タイプ']` の部分を先に計算しようとしてパニックになってしまうからである。

## データを並び変える

ポケモンスリープにおいて，「お手伝い時間が短い（速い）」ポケモンは非常に優秀である．
データ全体を，特定の列の数値を基準にして並べ替えるには，**`sort_values()`** というメソッドを使用する．

````
# お手伝い時間が短い（数値が小さい）順に並べ替える
df_sorted = df.sort_values('お手伝い時間(分)')

# 上位5匹のエリートポケモンを表示してみよう！
display(df_sorted.head())
````

このように，Pandasを使えば「ランキング」を作ることも一瞬である．

**【ワンポイントアドバイス：逆順にしたい場合】**
今回は「数値が小さい順（昇順）」にしたかったため，カッコの中に列名を書くだけでよかった．
もし逆に，「最大所持数が多い順」など「数値が大きい順（降順）」に並べ替えたい場合は，カッコの中に `ascending=False` というオプションを追加する．

````
# 参考：もし大きい順に並べたい場合の書き方
df_sorted_reverse = df.sort_values('お手伝い時間(分)', ascending=False)
display(df_sorted_reverse.head())
````

## NumPyで学んだ知識はPandasでどう活きるのか？

Pandasは，実は **NumPyのシステムを土台にして作られた拡張ツール** である．
そのため，前回までの講義で身につけたNumPyの知識やコードは，Pandasを使いこなす上でそのまま適用することができる．

具体的に，NumPyの操作がPandasでどう活きるのかを比較してみよう．

### 配列全体の計算（ブロードキャスト）

NumPyで学んだ「for文を使わずに，配列全体にまとめて同じ計算をする」という強力な機能は，PandasのSeries（1列のデータ）でも全く同じように使える．

````
# 【Pandasの場合】
# 「お手伝い時間」をすべて2倍にした新しい列データを作って付け加える
df['お手伝い時間の2倍'] = df['お手伝い時間(分)'] * 2
display(df)
````

列を消去するには`drop(columns=[消去したい列の名前])`を用いる．
以下のコードを実行すると元のDataFrameが得られる．

````
df = df.drop(columns=[`お手伝い時間の2倍`])
display(df)
````

### NumPyの数学関数をそのまま適用できる

Pandas自体にも平均などを求める機能はあるが，NumPyの `np.max()` や `np.mean()` などの関数に，Pandasの列を直接渡して計算させることも可能である．

````
import numpy as np

# NumPyの関数にPandasの列（Series）をそのまま渡せる
max_time = np.max(df['お手伝い時間(分)'])
print(f"最大のお手伝い時間は {max_time} 分である．")
````

### いつでもNumPyの2次元配列に戻せる

Pandasでデータを綺麗に整えたり条件で絞り込んだりした後，複雑な行列計算をするためにNumPyの配列に戻したいという場面がある．
その時は，`.values` または `.to_numpy()` というメソッドを使うだけで，NumPyの2次元配列（`np.ndarray`）に戻すことができる．

````
# PandasのDataFrameから，中身のデータだけをNumPy配列として取り出す
numpy_array = df.to_numpy()
print(numpy_array)
````

このように，Pandasは**NumPyの「圧倒的な計算力」に，Excelのような「表の扱いやすさ」を合体させた最強のツール**である．
