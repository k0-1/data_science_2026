## 行列の演算

### 行列同士の四則演算（要素ごとの計算）
NumPy配列同士の `+`, `-`, `*`, `/` は，**「同じ位置にある要素同士」** で計算されます．

````
A = np.array([[1, 2], 
              [3, 4]])
B = np.array([[10, 20], 
              [30, 40]])

print(A + B)  # 要素ごとの足し算
# 出力
# [[11 22]
#  [33 44]]
print(A * B)  # 要素ごとの掛け算（※行列の積ではありません！）
# 出力
# [[ 10  40]
#  [ 90 160]]
````

## 行列の積
線形代数で学ぶ「行と列の掛け合わせ（行列の積）」を行うには，`*` ではなく **`@` 演算子**，または **`np.dot()` 関数** を使用します．

行列 $A$ が $m \times k$ 行列、行列 $B$ が $k \times n$ 行列のとき，積$AB$が計算できて結果は $m \times n$ 行列になります．積が定義されるためには内側の次元（$k$）が一致している必要があります．

````
# 行列の積
print(A @ B)

# 関数を使う書き方（全く同じ結果になる）
print(np.dot(A, B))

# 出力
# [[ 70 100]
#  [150 220]]
# [[ 70 100]
#  [150 220]]
````

## スカラー倍
行列のスカラー倍とはすべての成分に同じ数を書けることをいう．
例えば$2A$は行列$A$のすべての成分を2倍した行列を表す．
この計算は，ブロードキャストを用いて次のように書ける．

````
print(2*A)
# 出力 
# [[2 4]
#  [6 8]]
````

### 行列とベクトルの積

$n$次元ベクトル$x$ (sizeが$n$のndarray) に$m\times n$の行列をかけることができる．
このとき，$x$は$n\times 1$行列と解釈される．

例えば$A x=\begin{pmatrix}1 & 2\\ 3 & 4\end{pmatrix}\begin{pmatrix}1\\2 \end{pmatrix}=\begin{pmatrix}5\\11 \end{pmatrix}$の計算は次のように書ける．

````
x = np.array([1,2])
print(A@x)
# 出力 [ 5 11]
````

## 行列に対する操作

### 転置行列（Transpose）
転置とは，行列の「行」と「列」を入れ替える操作のこと．
行列$A$を転置した行列を$A$の**転置行列**という．
行列の積を計算する際に，形状（Shape）を合わせる目的で非常によく使われる．
`.T` という属性を用いる．

````
C = np.array([[1, 2, 3], 
              [4, 5, 6]]) # 2行3列

# 転置して 3行2列 にする
print("元の行列:\n", C)
print("転置行列:\n", C.T)
# 出力: 
# 元の行列:
#  [[1 2 3]
#  [4 5 6]]
# 転置行列:
#  [[1 4]
#  [2 5]
#  [3 6]]
````

### 配列の結合（vstack, hstack）
バラバラのデータを縦や横にくっつける方法を見る．
データセットに新しい項目（列）を追加したり，新しいサンプル（行）を追加したりする実務で頻繁に使われる．

````
P = np.array([[1, 2], [3, 4]])
Q = np.array([[5, 6], [7, 8]])
# 縦に結合 (Vertical: Pの下にQをくっつける)
# ※結合する配列はタプル () で渡します
print(np.vstack((P, Q))) # 縦にP, Qの順で結合
print(np.vstack((P, Q, P))) # 縦にP, Q, Pの順で結合

# 横に結合 (Horizontal: Pの右にくっつける)
C = np.array([[10], [20]])
print(np.hstack((A, C)))
````

## ベクトルに対する操作 (内積とノルム)
ここからは1次元の配列（ベクトル）の計算に注目する．
AIが「この2つのデータは似ているか？」を判定するとき，
ベクトルの計算が行われる．

### ベクトルの内積
行列の積と同じ `@` を用いる．要素同士を掛けて，すべて足し合わせた「1つの数値」になる．

````
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("ベクトルの内積:", a @ b) 
# 出力 32 (=1*4 + 2*5 + 3*6)
````

### ベクトルのノルムと距離
2つのベクトルの「差」の長さを測ることで，データ間の「距離」を計算できる．
`np.linalg.norm()` を用いる．
値が小さいほど，近い点といえる．

````
dist = np.linalg.norm(a - b) #ベクトルaとbの距離
print(np.round(dist, 2))
````

### コラム： `linalg` って何？
`np.linalg.norm()` という少し長いコマンドが出てきた．
`import numpy as np` を実行した時点で，この機能はすでに使えるようになっている．

*   **`linalg` の意味**: これは **Lin**ear **Alg**ebra（線形代数）の略．
*   **なぜこんな書き方をするの？**: NumPyという巨大な道具箱の中には，無数の機能が入っている．そのため，「線形代数（`linalg`）」「乱数（`random`）」「統計（`stats`）」のように，機能ごとに「引き出し（サブモジュール）」が分かれている．
*   **読み方**: `.`（ドット）は「～の」と読む．つまり `np.linalg.norm` は，**「NumPy の 線形代数引き出し の ノルム計算」**を指定していることになる．

## コサイン類似度（Cosine Similarity）
内積とノルムを組み合わせると，AI分野で最も有名な類似度指標の一つである「コサイン類似度」が計算できる．これは2つのベクトルの「なす角」を評価するもので、文章の類似度（ChatGPTの裏側など）で毎日のように使われている（$1$に近いほど似ていて、$-1$に近いほど真逆）

$$ \text{コサイン類似度}=\cos \theta = \frac{(a,b)}{ \| a\| \|b\| }$$

$(a, b)$は$a$と$b$の内積，$\theta$は$a$, $b$のなす角である．

````
# コサイン類似度 = (aとbの内積) ÷ (aの長さ × bの長さ)
cos_sim = (a @ b) / (np.linalg.norm(a) * np.linalg.norm(b))
print("コサイン類似度:", np.round(cos_sim, 3))
````

### ? コラム：なぜ「距離」と「コサイン類似度」の2つを学んだの？
今回，データの似ている度合いを測る方法として，ベクトルの引き算（ユークリッド距離）と，角度を測る（コサイン類似度）の2つが登場した．これらはデータによって使い分ける．

*   **ユークリッド距離**：「数値の絶対的な大きさ」が重要なときに用いられる（例：位置情報，気温，金額）．
*   **コサイン類似度**：「大きさ」は無視して，「データの傾向（割合・パターン）」を見たいときに使われる．
